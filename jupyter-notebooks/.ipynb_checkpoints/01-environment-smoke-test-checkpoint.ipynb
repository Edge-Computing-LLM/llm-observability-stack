# Tutorial 01: Environment Smoke Test

This notebook verifies that your local **k3s** cluster, **NVIDIA GPU** runtime, and the **llm-observability-stack** pods are healthy and reachable.

### Prerequisites
- Cluster is running (`sudo k3s kubectl get nodes`)
- NVIDIA device plugin is installed
- This notebook is running with **Python 3.11** (`/usr/local/bin/python3.11`)

---

## 1. Verify Local Python Environment
Check that we are using the correct Python 3.11 interpreter.

```python
import sys
import os
print(f"Python Version: {sys.version}")
print(f"Executable: {sys.executable}")

assert "3.11" in sys.version, "Error: This notebook must run on Python 3.11"
```

## 2. Cluster Connectivity & Namespace Check
Verify that the `llm-observability` namespace exists and list the running pods.

```python
import subprocess

def run_k3s_cmd(args):
    cmd = ["sudo", "k3s", "kubectl"] + args
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0:
        print(f"Error: {result.stderr}")
    return result.stdout

print("Pods in 'llm-observability' namespace:")
print(run_k3s_cmd(["get", "pods", "-n", "llm-observability"]))
```

## 3. NVIDIA GPU Verification
Confirm that the cluster can see the GPU and it is allocatable.

```python
gpu_info = run_k3s_cmd(["get", "nodes", "-o", "jsonpath={.items[0].status.allocatable}"])
print(f"Node Allocatable Resources: {gpu_info}")

if "nvidia.com/gpu" in gpu_info or "nvidia.com/gpu.shared" in gpu_info:
    print("\nSUCCESS: NVIDIA GPU is detected and available in k3s!")
else:
    print("\nWARNING: No NVIDIA GPU detected in cluster allocatable resources.")
```

## 4. Service Accessibility
Check the internal services for Ollama, LangChain Demo, and Open WebUI.

```python
print(run_k3s_cmd(["get", "svc", "-n", "llm-observability"]))
```
