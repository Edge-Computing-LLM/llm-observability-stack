# Tutorial 08: Troubleshooting & etcd Simulations

Explore advanced support-engineering scenarios by using **etcd** to simulate backing-store failures.

---

## 1. Enabling etcd
The stack includes an `etcd` component that is disabled by default to save resources. To enable it for this tutorial:

```bash
helm upgrade llm-observability-stack . -f values.local-k3s.yaml -n llm-observability --set etcd.enabled=true
```

### Verification
Wait for the `etcd-0` pod to reach `Running` status.

````python
import subprocess
def run_k3s_cmd(args):
    return subprocess.run(["sudo", "k3s", "kubectl"] + args, capture_output=True, text=True).stdout

print("Checking etcd pod status:")
print(run_k3s_cmd(["get", "pods", "-n", "llm-observability", "-l", "app.kubernetes.io/name=etcd"]))
````

## 2. Simulate Connectivity Loss
We can simulate a failure of the backing store by scaling the etcd statefulset to zero and observing how the dependent services (like a future version of our proxy) might handle it.

**Command:**
```bash
kubectl scale statefulset etcd --replicas=0 -n llm-observability
```

### Observing the Impact
In a production environment, this would cause application timeouts or database errors. By simulating this in our k3s cluster, we can test our monitoring alerts and fallback logic.

## 3. Reviewing Proxy Logs
When things go wrong, the first place to look is the `langchain-demo` logs. We can use the LangSmith traces to correlate specific errors with user requests.

````python
print("Recent Proxy Logs:")
print(run_k3s_cmd(["logs", "deployment/langchain-demo", "-n", "llm-observability", "--tail=50"]))
````
